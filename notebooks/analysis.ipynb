```python
import pandas as pd
```

```python
df = pd.read_csv('../output/obiettori.csv')
```

```python
df
```

```python
# Filter out 2015 and 2016
df_filtered = df[~df['anno'].isin([2015, 2016])]

# Group by year and calculate totals
yearly_totals = df_filtered.groupby('anno').agg({
    'obiettori': 'sum',
    'totale': 'sum'
}).reset_index()

# Calculate the percentage
yearly_totals['percentuale_obiettori'] = (yearly_totals['obiettori'] / yearly_totals['totale'] * 100).round(1)

# Sort by year and display
yearly_totals = yearly_totals.sort_values('anno')
```
